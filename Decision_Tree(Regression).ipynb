## *Splitting Features and target Variable*


```python
X = Dataset[["age" , "income"]]
Y = Dataset["experience"]
```

## *Splitting Dataset into Training and Testing Sets*


```python
X_train , X_test , Y_train , Y_test = train_test_split(X , Y , random_state=47 , test_size=0.2)
```

## *Training the Decision Tree Regressor*


```python
df = DecisionTreeRegressor()
df.fit(X_train , Y_train)
```

```python
df.score(X_train , Y_train)*100
```

```python
df.score(X_test , Y_test)*100
```

## *Visualizing the Decision Tree Regressor*


```python
plot_tree(df)
plt.show()
```
