## *Filling Missing Age Values with Mean, Experience Mean, and Salary Mode*


```python
Dataset["Age"].fillna(Dataset["Age"].mean() , inplace=True)
Dataset["Age"].fillna(Dataset["Years of Experience"].mean() , inplace=True)
Dataset["Age"].fillna(Dataset["Salary"].mode()[0] , inplace=True)
```

```python
Dataset.dropna()
```

## *Splitting Features and target Variable*


```python
X = Dataset[["Age" ,"Years of Experience"]]
y = Dataset["Salary"]
```

## *Checking Missing Values in the Dataset*


```python
Dataset.isnull().sum()
```

## *Splitting Dataset into Training and Testing Sets*


```python
X_train , X_test , Y_train , Y_test = train_test_split(X , y ,random_state=42 , test_size=0.2)
```

## *Training K-Nearest Neighbors (KNN) Classifier*


```python
kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(X_train ,Y_train)
```
